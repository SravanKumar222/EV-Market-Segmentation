# **STATEWISE ELECTRIC , NON ELECTRIC VEHICLES**

```python
# EV Market Segmentation by Geography (State-wise)
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/RS_Session_260_AU_2349_A_to_B.csv")

# Drop Sl.No. (not required for analysis)
df = df.drop(columns=["Sl.No."])
df
```

## **EV PENETRATIOM**

```python
# Calculate Total Vehicles and EV Penetration Rate
df["Total"] = df["Electric"] + df["Non-electric"]
df["EV_Penetration_%"] = (df["Electric"] / df["Total"]) * 100

# Sort states by EV penetration
df_sorted = df.sort_values(by="EV_Penetration_%", ascending=False)

# Top 10 states by EV Penetration
top10 = df_sorted.head(10)

# Plot 1: EV Penetration (Top 10 states)
plt.figure(figsize=(12,6))
plt.bar(top10["State/UT"], top10["EV_Penetration_%"], color="green")
plt.xticks(rotation=45, ha="right")
plt.ylabel("EV Penetration (%)")
plt.title("Top 10 States/UTs by EV Penetration Rate")
plt.show()
```

# **EV VS NON-EV**

```python
# Plot 2: EV vs Non-EV Distribution (Top 5 states by total EVs)
top5_ev = df.sort_values(by="Electric", ascending=False).head(5)

top5_ev.plot(
    x="State/UT",
    y=["Electric", "Non-electric"],
    kind="bar",
    stacked=True,
    figsize=(12,6),
    color=["green", "gray"]
)
plt.ylabel("Number of Vehicles")
plt.title("Top 5 States by EV Adoption (Absolute Numbers)")
plt.xticks(rotation=45, ha="right")
plt.show()
```

# ** Key Insights:**
# 1. States like Delhi, Chandigarh, and Goa show higher EV penetration.
# 2. Large states like Uttar Pradesh and Maharashtra lead in total EV numbers.
# 3. Smaller states/UTs may be early adopters due to policy push and smaller fleets.
# 4. Targeting high-penetration states can accelerate EV adoption campaigns.
